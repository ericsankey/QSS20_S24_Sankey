## 1.3 sentiment analysis  (10 points)

A. Subset the press releases to those labeled with one of three topics via `topics_clean`: Civil Rights, Hate Crimes, and Project Safe Childhood. We'll call this `doj_subset` going forward and it should have 717 rows.



```python
## your code here for subsetting
doj_subset = doj[doj["topics_clean"].isin(['Civil Rights', 'Hate Crimes', 'Project Safe Childhood'])] 
doj_subset.shape
```

B. Write a function that takes one press release string as an input and:

- Removes named entities from each press release string (**Hint**: you may want to use `re.sub` with an or condition)
- Scores the sentiment of the entire press release using the `SentimentIntensityAnalyzer` and `polarity_scores`
- Returns the length-four (negative, positive, neutral, compound) sentiment dictionary (any order is fine)

Apply that function to each of the press releases in `doj_subset`. 

**Hints**: 

- A function + list comprehension to execute will takes about 30 seconds on a respectable local machine and about 2 mins on jhub; if it's taking a very long time, you may want to check your code for inefficiencies. If you can't fix those, for partial credit on this part/full credit on remainder, you can take a small random sample of the 717


```python
## your code here to define function
def prsa(string: str):
    """Press Release Sentiment Analyzer"""
    
    # Remove named entities
    named_entities = nlp(string)
    for entity in named_entities.ents:
        if entity.text.isalpha():
            string = re.sub(re.escape(entity.text), "", string)
    
    # Get sentiment scores
    sent_obj = SentimentIntensityAnalyzer()
    sentiment = sent_obj.polarity_scores(string)
    
    return sentiment
```

```python
## your code here executing the function
sentiment_scores_list = [prsa(press_release) for press_release in doj_subset.contents]

# Print the sentiment scores for each press release
for idx, sentiment_scores in enumerate(sentiment_scores_list):
    print(f"Press Release {idx + 1} Sentiment Scores:")
    print(sentiment_scores)
    print()
```

C. Add the four sentiment scores to the `doj_subset` dataframe to create a dataframe: `doj_subset_wscore`. Sort from highest neg to lowest neg score and print the top `id`, `contents`, and `neg` columns of the two most neg press releases. 

Notes:

- Don't worry if your sentiment score differs slightly from our output on GitHub; differences in preprocessing can lead to diff scores

```python
## your code here
# Assuming doj_subset is a DataFrame containing press releases

# Define a function to apply prsa function to each press release and return sentiment scores
def apply_prsa(contents):
    return prsa(contents)

# Apply the function to each press release in doj_subset
doj_subset['Sentiment_Scores'] = doj_subset['contents'].apply(apply_prsa)

# Expand the sentiment scores dictionary into separate columns
doj_subset[['Negative', 'Positive', 'Neutral', 'Compound']] = pd.DataFrame(doj_subset['Sentiment_Scores'].tolist(), index=doj_subset.index)

# Create a new dataframe with sentiment scores
doj_subset_wscore = doj_subset.copy()

# Sort the dataframe by negative sentiment score in descending order
doj_subset_wscore = doj_subset_wscore.sort_values(by='Negative', ascending=False)

# Print the top id, contents, and neg columns of the two most negative press releases
for idx, row in doj_subset_wscore.head(2).iterrows():
    print("Press Release ID:", row['id'])
    print("Contents:", row['contents'])
    print("Negative Score:", row['Negative'])
    print()
```

    - Only retains words 4 characters or longer
    - Uses the snowball stemmer from nltk to stem

- Returns a joined preprocessed string

id = 16-217 (this case: https://www.wlbt.com/story/32275512/three-mississippi-correctional-officers-indicted-for-inmate-assault-and-cover-up/)
    
**Resources**:

- Here's code examples for the snowball stemmer: https://www.geeksforgeeks.org/snowball-stemmer-nlp/

```python
custom_doj_stopwords = ["civil", "rights", "division", "department", "justice",
                        "office", "attorney", "district", "case", "investigation", "assistant",
                       "trial", "assistance", "assist"]
```

```python
## your code defining a text processing function
# Define Snowball stemmer
stemmer = SnowballStemmer('english')

# Define function for text preprocessing
def preprocess_text(raw_text):
    # Convert words to lowercase
    raw_text = raw_text.lower()
    
    # Remove stopwords
    stop_full = set(stopwords.words('english')).union(custom_doj_stopwords)
    text = ' '.join([word for word in word_tokenize(raw_text) if word not in stop_full])
    
    # Only retain alpha words and words 4 characters or longer
    text = re.sub(r'[^a-zA-Z]', ' ', text)
    text = ' '.join([word for word in word_tokenize(text) if len(word) >= 4])
    
    # Stem words
    text = ' '.join([stemmer.stem(word) for word in word_tokenize(text)])
    
    return text
```

```python
## your code executing the function
doj_subset_wscore['processed'] = [preprocess_text(text) for text in doj_subset_wscore['contents']]
doj_subset_wscore
```

```python
def create_dtm(list_of_strings, metadata):
    vectorizer = CountVectorizer(lowercase = True)
    dtm_sparse = vectorizer.fit_transform(list_of_strings)
    dtm_dense_named = pd.DataFrame(dtm_sparse.todense(), 
        columns=vectorizer.get_feature_names())
    dtm_dense_named_withid = pd.concat([metadata.reset_index(), dtm_dense_named], axis = 1)
    return(dtm_dense_named_withid)
```
