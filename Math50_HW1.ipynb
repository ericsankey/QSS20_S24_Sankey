```python
!pip install basemap
# Required libraries
import matplotlib.pyplot as plt
from mpl_toolkits.basemap import Basemap

# Coordinates for Hanover, NH (latitude, longitude)
hanover_lat = 43.7022
hanover_lon = -72.2896

# Create a map of the world
plt.figure(figsize=(12, 8))
m = Basemap(projection='mill', llcrnrlat=-60, urcrnrlat=90, llcrnrlon=-180, urcrnrlon=180, resolution='c')

# Draw coastlines, countries, and fill continents
m.drawcoastlines()
m.drawcountries()
m.fillcontinents(color='lightgray', lake_color='aqua')

# Draw the map boundaries and oceans
m.drawmapboundary(fill_color='aqua')

# Mark Hanover, NH with a red star
x, y = m(hanover_lon, hanover_lat)
m.plot(x, y, 'r*', markersize=15, label='Hanover, NH')

#part 3

salt_lake_city_lat = 40.7608
salt_lake_city_lon = -111.8910

w, h = m(salt_lake_city_lon, salt_lake_city_lat)
m.plot(w, h, 'g*', markersize=15, label='Salt Lake City, UT')


# Add a title and show the plot
plt.title('World Map with Hanover, NH and Salt Lake City, UT Marked by Stars')
plt.legend()
plt.show()
```
